### **d. Vẽ phổ điểm**

```python
def retrieve_data(df, column):
    new_df = df[column].value_counts().sort_index().reset_index()
    new_df.columns = ['Diem thi', 'So luong']
    return new_df
```

```python
def graphing(df, subject):
     plt.figure(figsize = (10, 10))

     plt.title(f'Biểu đồ phân phối điểm thi môn {subject} trong kỳ thi tốt nghiệp THPTQG 2022', fontdict = {
          'fontsize': 14,
          'fontweight': 'bold',
          'fontfamily': 'Ubuntu Medium'
     })

     plt.xlabel('Điểm thi')
     plt.xticks(rotation = 90)
     plt.ylabel('Số lượng')

     sns.barplot(data = df, x = 'Diem thi', y = 'So luong', palette = 'Blues')
```

#### **_i. Môn Toán_**

```python
diemtoan = retrieve_data(diemthi, 'toan')
graphing(diemtoan, 'Toán')
plt.savefig('Toan_2022.png')
```

### **Nhận xét**
- Mức điểm mà nhiều học sinh đạt được nhất là 7.8.
- Phần lớn học sinh đạt được mức điểm dao động trong khoảng từ 5.0 đến 8.0 --> Điều này chứng tỏ đề thi đảm bảo được điều kiện tốt nghiệp cho các thí sinh nếu các thí sinh có quá trình luyện tập từ trước
- Tuy nhiên, mức điểm từ 8.0 điểm trở xuống có số lượng thí sinh ít dần đi. Ở mức điểm trên 9.0, chỉ khoảng 15000 thí sinh đạt được mức điểm này so với tổng số gần 1 triệu thí sinh tham dự kì thi. --> Điều này chứng tỏ đề thi vẫn đảm bảo đủ độ khó để các thí sinh mong muốn điểm cao có thể đạt được và sử dụng nó để xét tuyển vào các trường Đại học bằng phương thức xét điểm thi THPTQG
