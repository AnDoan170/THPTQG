### **LOAD NECESSARY FILES**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')

gas
```

```python
plt.figure(figsize = (8,5))

plt.title('Gas prices over time in USD')

plt.plot(gas.Year, gas.USA, 'b.-')
plt.plot(gas.Year, gas.Canada, 'g.-')
plt.plot(gas.Year, gas['South Korea'], 'r.-')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.show()
```

```python
plt.figure(figsize = (8,5))

plt.title('Gas prices over time in USD', fontdict = {
    'fontweight': 'bold', 
    'fontsize': 22,
    'font': 'Cascadia Code'
})

for country in gas[1:]:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker = '.')

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.savefig('Gas_prices_over_time.png')
plt.show()
```

# FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa
```

### Histogram

```python
bins = [x for x in range(40, 101, 10)]

plt.hist(fifa.Overall, bins = bins, color = '#000000')

plt.xticks(bins)

plt.ylabel('Number of players')
plt.xlabel('Skill level')
plt.title('Distribution of FIFA player')

plt.show()
```

### Pie chart

```python
fifa['Preferred Foot'].sample(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f')

plt.show()
```
